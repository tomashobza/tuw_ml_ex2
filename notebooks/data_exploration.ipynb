# Data Exploration

## Housing Prices

```python
import pandas as pd
import os

hp_path = os.path.join('..', 'data', 'Housing.csv')
df_hp = pd.read_csv(hp_path)
```

```python
df_hp.shape
```

```python
df_hp.head()  
```

```python
# any missing values?
df_hp.isnull().sum()
```

```python
# dtypes correct?
df_hp.dtypes
```

```python
# check for outliers in numerical columns
numerical_cols = df_hp.select_dtypes(include=['float64', 'int64']).columns
df_hp[numerical_cols].describe()

# plot box plots for numerical columns
import matplotlib.pyplot as plt
import seaborn as sns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df_hp[col])
    plt.title(f'Box plot of {col}')
    plt.show()
```
